# Zipf’s Law

La ley de Zipf es una técnica utilizada para estudiar el comportamiento de las palabras en un cuerpo de texto. Con el estudio realizado por George Kingsley Zipf, se determinó que las palabras repetidas con mayor frecuencia cumplían un patrón en común. Este patrón se trata que, la primera palabra más repetida, es el doble de veces más usada que la segunda, el triple de veces más repetida que la tercera palabra y así sucesivamente.  
En la siguiente línea se realiza la importación del archivo zipfslaw, donde se encuentra todo el código utilizado para este proceso.

```python
#Importación del archivo sipfslaw que es el que contiene el código de funcionamiento
import zipfslaw
```

Lo que se realiza posteriormente, es la creación de la función principal (main), donde se van a llamar el resto de funciones creadas en el archivo zipfslaw.py. En este archivo se declaran las funciones para generar la tabla y luego para mostrarla. 

```python
def main():

    print("-----------------")
    print("| codedrome.com |")
    print("| Zipf's Law    |")
    print("-----------------\n")
 
    try:
        f = open("dracula.txt", "r", encoding="utf8")
        text = f.read()
        f.close()

        zipf_table = zipfslaw.generate_zipf_table(text, 135)

        zipfslaw.print_zipf_table(zipf_table)

    except IOError as e:

        print(e)
```

En la primera columna, que es el rango, se muestra la posición en la que está cada palabra. En la segunda columna, se escribe literalmente la palabra. En la tercera columna, se encuentra el número de veces que aparece la palabra en el texto. La fracción de Zzipf hace referencia a que la cantidad de veces que se repite la segunda palabra, es la mitad de la primera; la tercera, es el tercio de la primera y así sucesivamente. Para la frecuencia de Zipf, ya se aplica la fórmula de fracción de Zipf y se obtiene el resultado. En la diferencia, simplemente se resta la frecuencia actual con la frecuencia de Zipf y finalmente, se muestra este resultado en forma de porcentaje.
